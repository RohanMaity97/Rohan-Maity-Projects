# Data Cleaning

```python
#First, lets import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Now, we load the dataset and test to see if its has loaded as a dataframe and also check its shape. However, encoding issue occurs when trying to normally load it and thus, 'unicode_escape' is used 

df = pd.read_csv('Dataset.csv' , encoding='unicode_escape')
df
```

```python
df.shape
```

```python
# Lets check the first 10 values of the dataframe

df.head(10)
```

```python
# Now lets check the information of the dataframe to find its datatypes and see if it has any null values or not

df.info()
```

```python
# We find two blank/unrelated columns, so lets delete them

df.drop(['Status', 'unnamed1'], axis = 1, inplace=True)
```

```python
df.info()
```

```python
# Lets check the null values present in each of the columns

pd.isnull(df).sum()
```

```python
# Lets delete the null values present in the dataframe

df.dropna(inplace=True)
pd.isnull(df).sum()
```

```python
# Lets change the datatype of Amount column from float to integers

df['Amount'] = df['Amount'].astype('int')
df.info()
```

```python
# Lets rename the Cust_name into Customer_Name

df.rename(columns={'Cust_name':'Customer_Name'}, inplace = True)
df.columns
```

# Exploratory Data Analysis (EDA)

## Descriptive Statistics

```python
# Since the data has been cleaned and fixed, lets find out some descriptive statistical values of the dataset

df.describe()
```

```python
# Considering the fact that only Age, Orders, and Amount are the relevant numerical data, their descriptive statistical values can be filtered out

df[['Age','Orders','Amount']].describe()
```

## Gender Charts

```python
# let's plot a bar chart for the total count of shoppers by genders in the data

ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# let's plot a bar chart for the total amount generated from sales by each of the genders in the data

sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bx = sns.barplot(x='Gender', y='Amount', data=sales_gen)
for bars in bx.containers:
    bx.bar_label(bars)
```

## Age Charts

```python
# let's plot a bar chart for the age-group in the data

cx = sns.countplot(x='Age Group', data=df)
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
# let's plot a bar chart for the age-group in the data based on genders

dx = sns.countplot(x='Age Group', data=df, hue = 'Gender')
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
# let's plot a bar chart for the total amount generated from sales by each of the age groups in the data

sales_age =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ex = sns.barplot(x='Age Group', y='Amount', data=sales_age)
for bars in ex.containers:
    ex.bar_label(bars)
```

## State Charts

```python
# let's plot a bar chart for the total number of orders generated from top 5 states in the data

sales_state =df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
fx = sns.barplot(x='State', y='Orders', data=sales_state)
for bars in fx.containers:
    fx.bar_label(bars)
```

```python
# let's plot a bar chart for the total amount generated from top 5 states in the data

sales_state =df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
gx = sns.barplot(x='State', y='Amount', data=sales_state)
for bars in gx.containers:
    gx.bar_label(bars)
```

## Marital Status Charts

```python
# let's plot a bar chart for the total count of married(0) and unmarried(1) people in the data

hx = sns.countplot(x='Marital_Status', data=df)
for bars in hx.containers:
    hx.bar_label(bars)
```

```python
# let's plot a bar chart for the total amount generated from married(0) and unmarried(1) people based on genders

sales_status =df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ix = sns.barplot(x='Marital_Status', y='Amount', data=sales_status, hue = 'Gender')
for bars in ix.containers:
    ix.bar_label(bars)
```

## Occupation Charts

```python
# let's plot a bar chart for the shoppers based on their occupation

jx = sns.countplot(x='Occupation', data = df, order=df['Occupation'].value_counts().index)
for bars in jx.containers:
    jx.bar_label(bars)
```

```python
# let's plot a bar chart for the total amount generated from each of the occupations in the data

sales_occup =df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
kx = sns.barplot(x='Occupation', y='Amount', data=sales_occup)
sns.set(rc={'figure.figsize': (25, 5)})
```

## Product Category Charts

```python
# let's plot a bar chart for the top 10 product categories by count in the data

lx = sns.countplot(x='Product_Category', data = df, order=df['Product_Category'].value_counts().index[:10])
for bars in lx.containers:
    lx.bar_label(bars)
sns.set(rc={'figure.figsize': (15, 5)})
```

```python
# let's plot a bar chart for the top 10 highest amount of sales generated from each of the product categories in the data

sales_cat =df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
mx = sns.barplot(x='Product_Category', y='Amount', data=sales_cat)
for bars in mx.containers:
    mx.bar_label(bars)
sns.set(rc={'figure.figsize': (25, 5)})
```
